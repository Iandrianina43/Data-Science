```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
diamonds = pd.read_csv('diamonds.csv')
```

```python
diamonds.head()
```

```python
diamonds.info()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='carat',y='price',data=diamonds)
```

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='carat',y='price',data=diamonds,edgecolor=None,alpha=0.1)
```

```python
plt.figure(figsize = (12,8), dpi=72)
myhist = sns.histplot(x='price',data=diamonds, bins=50)

myhist.set_xlim(0, 18000)
myhist.set_ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(6, 4), dpi=72)
sns.countplot(data=diamonds, x='cut')
plt.show()
```

```python
plt.figure(figsize =(12,8), dpi=72)
sns.boxplot(x='cut', y='price', data=diamonds)
plt.show()
```

```python
# Code ici
plt.figure(figsize =(12,8), dpi=72)
sns.boxplot(x='cut', y='price', data=diamonds, order=['Fair', 'Ideal', 'Good', 'Very Good', 'Premium'], palette='blend:#45c5e4,#c545e4')
plt.show()
```
