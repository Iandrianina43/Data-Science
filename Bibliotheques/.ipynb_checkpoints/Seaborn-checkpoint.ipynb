```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
df.head()
```

```python
plt.figure(figsize =(10,10))
sns.scatterplot(x='salary',y='sales',data=df)
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='salary',y='sales',data=df, hue='division')
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x='salary',y='sales',data=df, hue='work experience')
```

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, hue='work experience')
```

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience')
```

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience', hue='work experience', palette='viridis', s=100)
```

```python
plt.figure(figsize =(12,8))
sns.scatterplot(x='salary', y='sales', data=df, style='work experience', hue='work experience', palette='viridis', s=100)
plt.savefig("plot.png")
```
