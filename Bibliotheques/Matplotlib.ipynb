```python
import numpy as np
m = np.linspace(0,10,11)
m
```

```python
c=3*10**8
c
```

```python
E=m*c**2
E
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(m,E)
plt.xlabel('masse (grammes)')
plt.ylabel('Energie (joules)')
plt.title('E=mc²')
plt.show()
```

```python
plt.plot(m,E)
plt.xlabel('masse(grammes)')
plt.ylabel('Energie(joules)')
plt.title('E=mc2')
plt.yscale('log')
plt.ylim(10**17,10**18)
plt.xlim(0,10)
plt.grid(True,which='minor',axis='y')
plt.show()
```

```python
labels = ['1 Mo','3 Mo','6 Mo','1 Yr','2 Yr','3 Yr','5 Yr','7 Yr','10 Yr','20 Yr','30 Yr']

july16_2007 =[4.75,4.98,5.08,5.01,4.89,4.89,4.95,4.99,5.05,5.21,5.14]
july16_2020 = [0.12,0.11,0.13,0.14,0.16,0.17,0.28,0.46,0.62,1.09,1.31]
```

```python
fig,ax1=plt.subplots()
```

```python
fig, ax1 = plt.subplots()

# Tracer les données sur le premier axe
ax1.set_xlabel("Mois")
ax1.set_ylabel("Date", color='red')
ax1.plot(labels, july16_2007, color='red')
ax1.tick_params(axis='y', labelcolor='red')

# Créer un second axe partageant le même axe x
ax2 = ax1.twinx()
ax2.set_ylabel("Date", color='blue')
ax2.plot(labels, july16_2020, color='blue')
ax2.tick_params(axis='y', labelcolor='blue')

# Afficher le graphique
plt.show()
```
